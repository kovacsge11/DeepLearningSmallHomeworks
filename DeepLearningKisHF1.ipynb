## Tasks 5 and 6

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from sklearn.preprocessing import StandardScaler

%matplotlib inline

#Lists created to contain the given channels for every picture
red_list = []
green_list = []
blue_list = []

#For cycle for iterating over all of the 5 pictures
for i in range(1,6):
    
    #Reading image
    img = mpimg.imread("%d.png" % (i))
    #Showing it
    plt.imshow(img)
    plt.show()

    #Printing the R, G and B channel values of the 32x32 picture
    #And adding them to the corresponding lists
    red = img[:,:,0]
    red_list.append(red)
    
    green = img[:,:,1]
    green_list.append(green)
    
    blue = img[:,:,2]
    blue_list.append(blue)
    
    print("Red values:\n" + str(red) + "\nGreen values:\n" + str(green) + "\nBlue values:\n" + str(blue))
    

#Turning the list of arrays into an array of size 32x32x5 for easier mean and deviation calculation
red_array = np.dstack(red_list)
green_array = np.dstack(green_list)
blue_array = np.dstack(blue_list)

#Created for better printing
channel_name_dict = {'red': red_array, 'green': green_array, 'blue': blue_array}

#Standard scaling all values in each pixel - channel pair for the 5 pictures
#And then printing the mean, deviation, and the scaled matrix
for key in channel_name_dict:
    channel = channel_name_dict.get(key)
    for row_pixel in range(0,32):
        for col_pixel in range(0,32):
            scaler = StandardScaler()
            channel[row_pixel, col_pixel,:] = scaler.fit_transform(channel[row_pixel, col_pixel,:].reshape(-1,1)).reshape(1,-1)
            print("Channel %s mean for pixel %d,%d: " % (key,row_pixel,col_pixel) + str(scaler.mean_))
            print("Channel %s variance for pixel %d,%d: " % (key,row_pixel,col_pixel) + str(scaler.var_))

print("Scaled red channel values for every pixel: " + str(red_array))
print("Scaled green channel values for every pixel: " + str(green_array))
print("Scaled blue channel values for every pixel: " + str(blue_array))
```
